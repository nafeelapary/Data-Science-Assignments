3. Who scores the most on average for math, reading and writing based on Test preparation course

4. What do you think about the scoring variation for math, reading and writing based on Gender

```python
plt.figure(figsize=(15,7))
plt.subplot(1,3,1)
sns.boxplot(x='gender',y='math score',data=marks)
plt.subplot(1,3,2)
sns.boxplot(x='gender',y='reading score',data=marks)
plt.subplot(1,3,3)
sns.boxplot(x='gender',y='writing score',data=marks)
plt.show()
```

In math,males have scored more than females. A female has scored zero marks in it. In reading and writing females have scored more.

4. What do you think about the scoring variation for math, reading and writing based on Test preparation course

```python
plt.figure(figsize=(15,7))
plt.subplot(1,3,1)
sns.boxplot(x='test preparation course',y='math score',data=marks)
plt.subplot(1,3,2)
sns.boxplot(x='test preparation course',y='reading score',data=marks)
plt.subplot(1,3,3)
sns.boxplot(x='test preparation course',y='writing score',data=marks)
plt.show()
```

The variation of scores is entirely different for those who have completed the course and others. The students who have completed test preparation course have scored very good marks.

5. The management needs your help to give bonus points to the top 25% of students based on their math score, so how will you help the management to achieve this.

```python
maxmark = marks[marks['math score']>=marks['math score'].quantile(0.75)]
maxmark
```
